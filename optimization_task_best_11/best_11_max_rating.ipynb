Задача:


Сформувати склад з 11 футболістів. Склад потрібно вибрати таким чином щоб сума рейтингів гравців на тих позиціях, на які ми їх ставимо, була максимальною. Наприклад, якщо ми ставимо Messi, Lionel на позицію AML, а Silva, Bernardo - на позицію AMR, сумарний рейтинг буде 90.57%+94.34%.
Також є обмеження - в склад можна взяти тільки по одному гравцю кожної національності (у випадку подвійної національності береться перша) та по одному гравцю з кожного клубу.
Тобто всі 11 футболістів в складі мають бути різних національностей та з різних клубів. 

```python
import pandas as pd
import numpy as np
import pulp as pl
```

```python
df = pd.read_excel('Test_Task_FM24.xlsx', 
                   sheet_name='Fantasy_FM24',
                   engine='openpyxl')
df.head()
```

```python
roles_dict = {
    'GK': 1,
    'DL': 1,
    'DC': 2,
    'DR': 1,
    'MC': 2,
    'AML': 1,
    'AMR': 1,
    'AMC': 1,
    'ST': 1
}
```

```python
df['Nation'] = df['Nation'].str.split('/').str[0].str.strip()
df['Club'] = df['Club'].replace('-', pd.NA)
```

```python
col_list = ['GK Rating', 'DL Rating', 'DC Rating', 'DR Rating', 'DM Rating', 'ML Rating', 'MC Rating', 'MR Rating', 'AML Rating', 'AMC Rating', 'AMR Rating', 'FS Rating', 'TS Rating']

for col in col_list:
    df[col] = df[col].str.rstrip('%').astype(float)

df.head()
```

```python
df['ST Rating'] = np.maximum(df['FS Rating'],df['TS Rating'])
df.head()
```

```python
rating_cols = [f"{role} Rating" for role in roles_dict]

id_vars = ['Unique ID', 'Name', 'Nation', 'Club']

df_melted = (df.melt(id_vars=id_vars,
                  value_vars=rating_cols,
                  var_name='role_col', value_name='rating')
            .assign(role=lambda x: x['role_col'].str.replace(' Rating', '', regex=False))
            .drop(columns='role_col'))
df_melted.head()
```

```python
prob = pl.LpProblem("Best11_MaxRating", pl.LpMaximize)

rows = list(df_melted.index)

x = {r: pl.LpVariable(f"x_{r}", lowBound=0, upBound=1, cat="Binary") for r in rows}

prob += pl.lpSum(df_melted.at[r, 'rating'] * x[r] for r in rows), "TotalRating"

# Додавання умов:
# 1. Ліміти по ролях
for role, cnt in roles_dict.items():
    role_rows = [r for r in rows if df_melted.at[r, 'role'] == role]
    prob += pl.lpSum(x[r] for r in role_rows) == cnt, f"Count_role_{role}"

# 2. Кожен гравець не більше 1 разу
for uid, idxs in df_melted.groupby('Unique ID').groups.items():
    prob += pl.lpSum(x[r] for r in idxs) <= 1, f"Uniq_player_{uid}"

# 3. Кожна нація не більше 1 гравця
for nat, idxs in df_melted.groupby('Nation').groups.items():
    prob += pl.lpSum(x[r] for r in idxs) <= 1, f"Uniq_nation_{nat}"

# 4. Кожен клуб не більше 1 гравця
for club, idxs in df_melted.groupby('Club', observed=True).groups.items():
    prob += pl.lpSum(x[r] for r in idxs) <= 1, f"Uniq_club_{club}"

status = prob.solve(pl.PULP_CBC_CMD(msg=True))  

print("Status:", pl.LpStatus[prob.status])

sel_rows = [r for r in rows if x[r].value() == 1]
solution = (df_melted.loc[sel_rows]
            .rename(columns={'rating':'Rating'})
            .sort_values(['role','Rating'], ascending=[True, False]))

print("\n=== Обраний склад ===")
print(solution.to_string(index=False))
print(f"\nСумарний рейтинг: {solution['Rating'].sum():.2f}")
```
